# Load terminology database

```python
#@title **The Pantry** (Terminology Database)
#@markdown #### <font color="orange">**NOTE:**</font> **Run the this cell to initialize the database.**

class log:
    f = lambda color: lambda string: print(color + string + "\33[0m")
    black = f("\33[30m")
    red = f("\33[31m")
    green = f("\33[32m")
    yellow = f("\33[33m")
    blue = f("\33[34m")
    megenta = f("\33[35m")
    cyan = f("\33[36m")
    white = f("\33[37m")

!wget -q --show-progress --no-cache --backups=1 'https://raw.githubusercontent.com/WASasquatch/noodle-soup-prompts/main/nsp_pantry.py'

import nsp_pantry
from nsp_pantry import nspterminology, nsp_parse

terminology_database = nspterminology

if terminology_database:
	log.green("Loaded terminology database from the pantry.")
	print("\x1B[3mMmm. Noodle Soup.\x1B[0m")
else:
	log.red("Unable to load terminology database")
```

# 2) Raw Noodle Prompts

```python
text_prompts = {
    0: [
        "Portrait of a _adj-beauty_ _noun-emote_ _nationality_ woman from _pop-culture_ in _pop-location_ with pearlescent skin and white hair by _artist_, _site_.:5",
        "_hd_, _hd_, _3d-terms_, _3d-terms_:2",
        "_color_ Color Scheme",
        ],
}
```

# Process Dynamic Text Prompts

```python
#@title <font size="">3) Cook up some Noodle Soup</font>
#@markdown *Run this cell to return your processed code to copy to Disco Diffusion. Dish up!*
import random

#@markdown ---
bowls_of_soup = 5 #@param {type: 'number'}
#@markdown &nbsp;&nbsp;&nbsp;&nbsp;**bowls_of_soup:** *How many prompt iterations should we run?*

def process_noodle_soup(text_prompts):
  terms = []
  for term in terminology_database:
    if term not in terms:
      terms.append(term)

  processed_prompt_list = {}
  processed_prompts = []

  print("")
  print("text_prompts = {")
  for pstep, pvalue in text_prompts.items():
    print("    "+str(pstep)+": [")
    for prompt in pvalue:
      new_prompt = prompt
      for term in terms:
        tk = '_'+term+'_'
        tc = prompt.count(tk)
        for i in range(tc):
          new_prompt = new_prompt.replace(tk, random.choice(terminology_database[term]), 1)
      processed_prompts.append(new_prompt)
    processed_prompt_list[pstep] = processed_prompts
    for np in processed_prompts:
      log.yellow("        \""+np+"\",")
    print("        ],")
    processed_prompts = []
  print("}")


log.green("TEXT PROMPTS CODE:")
for i in range(1, bowls_of_soup+1):
  print("")
  log.blue("\033[1mTEXT PROMPT "+str(i)+":\033[0m")
  process_noodle_soup(text_prompts)
```
